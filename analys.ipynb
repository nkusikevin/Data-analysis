```python
import numpy as np
import pandas as pd
```

```python
#load data files
data_file = pd.read_csv("data_with_headers.csv")
```

```python
print(data_file[0:3])
```

```python
time  = data_file['time']
```

```python
sensors = data_file.loc[:,'s1':'s4']
print(sensors[0:6])
```

```python
time = time - time[0]
print(time[0:3])
```

```python
avg_row = np.mean(sensors,1)
avg_col = np.mean(sensors,0)
print(avg_row[0:3])
print(avg_col)
```

```python
my_data = [time,sensors,avg_row]
result = pd.concat(my_data,axis=1)
print(result[0:3])
```

```python
result.to_csv("result.csv")
result.to_excel("result.xlsx")
result.to_html("result.html")
result.to_clipboard()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.figure(1)
plt.plot(time,sensors['s1'],'r-1')
plt.plot(time,avg_row, 'b.')
plt.legend(['Sensor 1' , 'Average'])
plt.xlabel("Time (sec)")
plt.ylabel("Sensor Values")
plt.savefig('myplot.png')
```
